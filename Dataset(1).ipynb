```
import zipfile
zip_ref = zipfile.ZipFile("/content/drive/My Drive/flickrfaceshq-dataset-ffhq.zip", 'r')
zip_ref.extractall("/content/ffhq")
zip_ref.close()
```

```
import tensorflow as tf
import utils
from models.Face_recognition import FR_model
from PIL import Image
import os
import argparse
import cv2


l = 0.5
r = 0.5

model=tf.keras.models.load_model('checkpoint2')

FR=FR_model()

os.mkdir("/content/drive/My Drive/520Dataset")
os.mkdir("/content/drive/My Drive/520Dataset/Fat")
os.mkdir("/content/drive/My Drive/520Dataset/Thin")
fat = 0
thin = 0
subdirs = os.listdir("/content/ffhq")


for file in subdirs:
    
    

  img = cv2.imread("/content/ffhq"+'/'+file)
  #print("/content/drive/My Drive/ffhq-dataset/images1024x1024/"+subdir+'/'+file)

  img_1 = img.copy()
  img_1 = cv2.cvtColor(img_1, cv2.COLOR_BGR2RGB)
  img_1 = cv2.resize(img, (128, 128))

  mean = img_1.mean()
  std = img_1.std()

  img_1 = (img_1 - mean)/std
  img_1 = img_1.reshape((1, 128, 128, 3))
  embedding = FR(img_1)
  pred = model.predict(embedding)
  #print("####", pred[0] > 0.5)
  if pred[0] < l:
      cv2.imwrite("/content/drive/My Drive/520Dataset/Fat/"+file, img)
      fat += 1
      print("fat", fat)
  elif pred[0] > r:
      cv2.imwrite("/content/drive/My Drive/520Dataset/Thin/"+file, img)
      thin += 1
      print("thin", thin)





```
